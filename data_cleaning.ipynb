```python
# function for cleaning the price column
def cl_price(in_put):
    # returning the float value if present
    if type(in_put) == float:
        return in_put
    else:
        # converting the lac into crore
        if in_put[1] == 'Lac':
            return round((float(in_put[0])/100),2)
        else:
            
            return float(in_put[0])
```

```python
data['price'] = data['price'].str.split().apply(cl_price)
```

```python
data.rename(columns = {'area_per_sqft':'price_per_sqft'} , inplace = True)
```

```python
# cleaning the price_per_sqft cloumn
def clean(in_put):
    in_put = str(in_put)
    return float(in_put.replace('₹ ','').replace(',' , '').split('/')[0])
    
```

```python
data['price_per_sqft'][0].replace('₹ ','').split('/')[0].replace(',' , '')
```

```python
data['price_per_sqft']  = data['price_per_sqft'].apply(clean)
```
